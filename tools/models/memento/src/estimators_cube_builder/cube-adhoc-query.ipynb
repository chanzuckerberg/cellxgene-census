# Query the estimators cube

```python
# load cube data

import json
from os import path
import tiledb as tdb

cube_dir = '/mnt/census/estimators-cube-70a7705-float32'
estimators_array = tdb.open(path.join(cube_dir, "estimators"), config={"soma.init_buffer_bytes": 2**32})
obs_groups_df = tdb.open(path.join(cube_dir, "obs_groups")).df[:].set_index('obs_group_joinid')
with open(path.join(cube_dir, "feature_ids.json"), 'r') as json_file:
    features = json.load(json_file)
```

```python
# filter to a single tissue and a few features, as an example

obs_groups_df = obs_groups_df[obs_groups_df.tissue_general_ontology_term_id.isin(['UBERON:0000970'])]

features = features[:3]
```

```python
# query for matching estimators and create the cube from the obs dimensions and estimator values

estimators = estimators_array.df[features, obs_groups_df.index.values].set_index('obs_group_joinid')
cube = obs_groups_df.join(estimators).reset_index().set_index(['feature_id', 'obs_group_joinid'])
cube
```
